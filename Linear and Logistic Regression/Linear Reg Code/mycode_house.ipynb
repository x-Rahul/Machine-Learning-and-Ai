```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Resources/Linear Regression/House_Price.csv', header=0)
df.head()
```

### Scatter Plot for Hotel Rooms vs Price 

```python
plt.scatter(x = df.n_hot_rooms, y = df.price)
plt.xlabel('Hotel Rooms')
plt.ylabel('Price')
```

Outliers present at Hotel room = 80 and 100

### Scatter plot Rainfall vs Price

```python
plt.scatter(df.rainfall, df.price)
plt.xlabel('Rainfall')
plt.ylabel('Price')
```
