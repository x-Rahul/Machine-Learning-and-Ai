## Correlation
Correlation is a statistical measure that indicates the extent to which two or more variables fluctuate together. A positive correlation indicates the extent to which those variables increase or decrease in parallel; a negative correlation indicates the extent to which one variable increases as the other decreases.<br>
We make corelation matrix to se correlation between variables. High corelation leads to multi-colinearity. We take the variable that has higher corelation value in the matrix to avoid multi-colinearity.<br>
Near 0: No corelation<br>
Near 1: High Corelation<br>
Near -1: High Negative Coreation<br>

```python
df.corr()
```

### Creating the Model
#### M1 : From statsmodels

```python
import statsmodels.api as sn
```

```python
X = sn.add_constant(df['room_num']) # stats does not include constant term. So adding it to data and creating variable X
```

```python
# Fit the Model
lm = sn.OLS(df['price'], X).fit() # OLS: Ordinary least Square.
```

```python
lm.summary()
```

#### M2 : sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
lm2 = LinearRegression()
```

```python
X = df[['room_num']]
y = df['price']
```

```python
lm2.fit(X,y)
# help(lm2) # gets documentation
```

```python
print(lm2.intercept_, lm2.coef_)
```

It is same as output from M1. We have y = mX + C.<br>
